# Parte 2: Seleção e filtros

## Selecionando as colunas que indicam o tipo de combustível e o valor médio

```python
df_tipo_comb_valor_medio = (
    df_gasp_brazil
    .rename(columns={
        "PRODUTO": "TIPO_COMBUSTIVEL", 
        "PREÇO MÉDIO REVENDA": "VALOR_MEDIO"
    })
    [["TIPO_COMBUSTIVEL", "VALOR_MEDIO"]]
)
df_tipo_comb_valor_medio.head()
```

## Filtrando registro que o preço da gasolina foi maior que R$4,00

```python
df_valor_maior_4 = (
    df_tipo_comb_valor_medio[df_tipo_comb_valor_medio["VALOR_MEDIO"] > 4]
)

display(df_valor_maior_4)

num_rows = df_valor_maior_4.shape[0]

print(f"Houveram {num_rows} ocorrências nesse filtro.")
```
